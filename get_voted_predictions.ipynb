```python
import pandas as pd

models = [
    "vit_h_14_pre-None_cls-one_hot_seed-0_e-15",
    "vit_h_14_pre-None_cls-one_hot_seed-42_e-15",
    "vit_h_14_pre-None_cls-one_hot_seed-1024_e-15",
    "vit_h_14_pre-None_cls-one_hot_seed-1729_e-15",
    "vit_h_14_pre-None_cls-one_hot_seed-2048_e-15"
]

save_name = "vit_h-one_hot-maj_vote_predictions.csv"

initial_predictions = pd.read_csv("./models/vit_h_14_pre-None_cls-one_hot_seed-0_e-15/predictions.csv")

labels = initial_predictions["annotation_id"]

prediction_vote_dict = {}

for label in labels:
    prediction_vote_dict[label] = {}

for i, model in enumerate(models):
    prediction_df = pd.read_csv(f"./models/{model}/predictions.csv")
    # Turn prediction df into a dict with keys annotation_id and values as concept_name
    prediction_dict = {}
    for index, row in prediction_df.iterrows():
        prediction_dict[row["annotation_id"]] = row["concept_name"]

        assert row["annotation_id"] in prediction_vote_dict, \
            f"annotation_id {row['annotation_id']} not found in prediction_vote_dict"
        
        if row["concept_name"] not in prediction_vote_dict[row["annotation_id"]]:
            prediction_vote_dict[row["annotation_id"]][row["concept_name"]] = 0
        prediction_vote_dict[row["annotation_id"]][row["concept_name"]] += 1

print("prediction_vote_dict", prediction_vote_dict)

majority_vote_dict = {}

for annotation_id in prediction_vote_dict:
    # Get the concept_name with the most votes
    # Check which ones have even split
    max_votes = 0
    max_concept_name = None
    for concept_name, votes in prediction_vote_dict[annotation_id].items():
        if votes > max_votes:
            max_votes = votes
            max_concept_name = concept_name

    majority_vote_dict[annotation_id] = max_concept_name

print(majority_vote_dict)

# Save majority_vote_dict to a csv file
majority_vote_df = pd.DataFrame.from_dict(majority_vote_dict, orient="index", columns=["concept_name"])

majority_vote_df.index.name = "annotation_id"

majority_vote_df.reset_index(inplace=True)
majority_vote_df.to_csv("processed_predictions/"+save_name, index=False)
```
