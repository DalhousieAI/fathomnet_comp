```python
def assign_taxonomic_ranks(row, taxonomic_concept_dict):
    """
    Assign taxonomic ranks for a given row using the taxonomic_concept_dict.
    If a rank is unavailable, assign the last known rank.
    """
    concept_name = row["label"]
    if concept_name in taxonomic_concept_dict:
        taxon_info = taxonomic_concept_dict[concept_name]
        last_known_rank = None
        for level in levels:
            if taxon_info[level] is not None:
                row[level] = taxon_info[level]
                last_known_rank = taxon_info[level]
            else:
                row[level] = last_known_rank
    else:
        # If concept_name is not in the dictionary, assign None to all levels
        for level in levels:
            row[level] = None
    return row

test_labels = test_labels.apply(assign_taxonomic_ranks, axis=1, taxonomic_concept_dict=taxonomic_concept_dict)
```
